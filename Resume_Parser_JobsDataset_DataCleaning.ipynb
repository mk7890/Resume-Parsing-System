2. Remove Special Characters & Extra Whitespaces.
- Cleaning text helps improve model performance.

```python
import re

def clean_text(text):
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = re.sub(r'[^a-zA-Z0-9,.\- ]', '', text)  # Remove special characters except commas, periods, hyphens
    return text.strip()

for col in ['Name', 'Title', 'Role', 'Qualifications', 'Skills', 'Company']:
    jobs_df[col] = jobs_df[col].apply(clean_text)
```

Reduce Dataset to 100 entries per unique Job Role

```python
# Select 100 rows for each unique job role
jobs_sampled_df = jobs_df.groupby('Role').apply(lambda x: x.sample(n=min(100, len(x)), random_state=42)).reset_index(drop=True)

# Check the size of the new dataframe
print(f"Reduced dataset size: {jobs_sampled_df.shape}")
```

3. Tokenization & Text Formatting for DistilBERT
- DistilBERT requires input in a specific tokenized format. Use Hugging Face's tokenizer:

```python
from transformers import DistilBertTokenizer

# Load DistilBERT tokenizer
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')

# Tokenize relevant columns
def tokenize_text(text):
    return tokenizer(text, padding='max_length', truncation=True, max_length=128, return_tensors="pt")

jobs_df['Title_tokenized'] = jobs_df['Title'].apply(tokenize_text)
jobs_df['Role_tokenized'] = jobs_df['Role'].apply(tokenize_text)
jobs_df['Qualifications_tokenized'] = jobs_df['Qualifications'].apply(tokenize_text)
jobs_df['Skills_tokenized'] = jobs_df['Skills'].apply(tokenize_text)
```

```python
# List of columns to tokenize
text_columns = ['Title', 'Role', 'Qualifications', 'Skills', 'Company', 'Name']

# Tokenize each column
for col in text_columns:
    jobs_df[f'{col}_tokenized'] = jobs_df[col].apply(lambda x: tokenizer(x, padding='max_length', truncation=True, max_length=128, return_tensors="pt"))
```
